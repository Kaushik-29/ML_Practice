Next we do
 1. Data Loading and Overview
 for that we use pandas to load csv file and to display the content in the csv file

```python
print("Loading the food waste dataset")
df = pd.read_csv("food_waste_data.csv")
print("First 5 rows:")
print(df.head())
print("Number of rows:", len(df))
print("Number of columns:", len(df.columns))
print("Shape:", df.shape)
print(list(df))
```

  2. Descriptive Statistics

```python
numeric_df = df.select_dtypes(include='number')

print("Descriptive Statistics:")
print(numeric_df.describe())

print("\nSum:")
print(numeric_df.sum())

print("\nStandard Deviation:")
print(numeric_df.std())

print("\nQuartiles:")
print(numeric_df.quantile([0.25, 0.5, 0.75]))

print("\nMinimum Values:")
print(numeric_df.min())

print("\nMaximum Values:")
print(numeric_df.max())


# Instead of calling sum(), std(), min(), max(), and quantile() separately,
# we can use df.describe() to get a complete statistical summary for all numeric columns.
```

3. Missing Values

```python
print("Missing Values ")
print(df.isnull().sum())
```

 4. Data Types and Type Inference

```python
print(" Data Types")
print(df.dtypes)
print("Data Info")
df.info()
```

  5. Unique and Frequency Counts

```python
print("Unique Values and Frequency Counts")
for col in df.columns:
    print(f"\nColumn: {col}")
    print("Unique:", df[col].nunique())
    print("Most frequent values:")
    print(df[col].value_counts().head())
```

6. Correlation and Covariance

Correlation = How strongly two variables are related, on a standard scale.

Covariance = Measures how much two variables change together

```python
print("Correlation Matrix:")
print(df.corr(numeric_only=True))

print("\nCovariance Matrix:")
print(df.cov(numeric_only=True))
```

7.   Data Visualisation:


    *   Bar Graph





```python
meal_waste = df.groupby('Meal_Type')['Total_Waste_Kg'].sum()

# Creating bar chart
plt.figure(figsize=(8, 5))
plt.bar(meal_waste.index, meal_waste.values, color=['skyblue', 'orange', 'lightgreen'])
plt.title("Total Waste per Meal Type")
plt.xlabel("Meal Type")
plt.ylabel("Total Waste (Kg)")
plt.tight_layout()
plt.show()
```


      *   Line Plot




```python
plt.figure(figsize=(10, 5))

for meal in df['Meal_Type'].unique():
    # Filter data for that meal type
    meal_data = df[df['Meal_Type'] == meal]
    plt.plot(meal_data['Date'], meal_data['Total_Waste_Kg'], marker='o', label=meal)

plt.title("Food Waste Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Total Waste (Kg)")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load the CSV file
df = pd.read_csv("/content/food_waste_data.csv")

cols = ['Extra_Food_Kg', 'Leftover_Food_Kg', 'Total_Waste_Kg']
# Basic Statistics
for col in cols:
    print(f"\n--- {col} ---")
    print(f"Mean: {df[col].mean():.2f}")
    print(f"Median: {df[col].median():.2f}")
    print(f"Variance: {df[col].var():.2f}")
    print(f"Standard Deviation: {df[col].std():.2f}")

# Quartiles
for col in cols:
    print(f"\n--- Quartiles for {col} ---")
    print(f"Q1 (25%): {df[col].quantile(0.25)}")
    print(f"Q2 (Median): {df[col].quantile(0.50)}")
    print(f"Q3 (75%): {df[col].quantile(0.75)}")

# Plot PDF
plt.figure(figsize=(14, 5))
for i, col in enumerate(cols):
    plt.subplot(1, 3, i+1) # Adjusted subplot to 3 columns
    sns.kdeplot(df[col], fill=True)
    plt.title(f"PDF of {col}")
plt.tight_layout()
plt.show()

# Plot CDF
plt.figure(figsize=(14, 5))
for i, col in enumerate(cols):
    plt.subplot(1, 3, i+1) # Adjusted subplot to 3 columns
    sorted_data = np.sort(df[col])
    yvals = np.arange(len(sorted_data)) / float(len(sorted_data) - 1)
    plt.plot(sorted_data, yvals, marker='o')
    plt.title(f"CDF of {col}")
    plt.xlabel(col)
    plt.ylabel('Probability')
plt.tight_layout()
plt.show()

# 5. Covariance and Correlation
print("\nCovariance Matrix:")
print(df[cols].cov())

print("\nCorrelation Matrix:")
print(df[cols].corr())
```
